```python
#W6A5
l=list(map(int,input().split()))
d={'positives':[],'negatives':[],'zeros':[]}
for i in l:
    if i>0:
        d.setdefault('positives',[]).append(i)
    elif i<0:
        d.setdefault('negatives',[]).append(i)
    elif i==0:
        d.setdefault('zeros',[]).append(i)
print(d)
```

```python
#W6A9
a=input().split()
b=input().split()
d1={}
d2={}
for i in a:
    x,y=i.split(':')
    d1.setdefault(x,[]).append(y)
for j in b:
    z,w=j.split(':')
    d2.setdefault(z,[]).append(w)
if x==z:
    d1.setdefault(x,[]).append(w)
else:
    d1.setdefault(z,[]).append(w)
print(d1)
```

```python
#W6A10
l=list(map(int,input().split()))
k=int(input())
n=[]
for i in l:
    if k-i in l and i<=k-i and (i,k-i) not in n:
        n.append((i,k-i))
    if k-i in l and i>=k-i and (k-i,i) not in n:
        n.append((k-i,i))
    if i==k-i and l.count(i)==1:
        n.remove((i,k-i))
print(n)
```
