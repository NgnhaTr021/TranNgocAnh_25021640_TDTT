```python
#W5A1
def Max_Of_Two(a,b):
    if a>b:
        return a
    if a<b:
        return b
a,b=map(int,input().split())
print(Max_Of_Two(a,b))
```

```python
#W5A2
def Swap(a,b):
    a,b=b,a
    return a,b
a,b=map(int,input().split())
print(Swap(a,b))
```

```python
#W5A3
flag=True
n=int(input())
for i in range(2,n):
    if n%i==0:
        flag=False
if flag:
    print("Yes")
else:
    print("No")
```

```python
#W5A4
n=int(input())
s=0
for i in range(1,n):
    if n%i==0:
        s+=i
if s==n:
    print("Yes")
else:
    print("No")
```

```python
#W5A7
num1,num2=map(float,input().split())
x=input()
if x=='+':
    print(f'{(num1+num2):.2f}')
if x=='-':
    print(f'{(num1-num2):.2f}')
if x=='*':
    print(f'{(num1*num2):.2f}')
if x=='/':
    print(f'{(num1/num2):.2f}')
```

```python
#W5A8
x,y=map(int,input().split())
a=bin(x)[2:]
b=bin(y)[2:]
n=0
if len(str(a))>len(str(b)):
    x=len(str(a))-len(str(b))
    for i in range(len(str(b))):
        if a[i]!=b[i]:
            n+=1
    n+=x
if len(str(b))>len(str(a)):
    x=len(str(b))-len(str(a))
    for i in range(len(str(a))):
        if a[i]!=b[i]:
            n+=1
    n+=x
print(n)
```
