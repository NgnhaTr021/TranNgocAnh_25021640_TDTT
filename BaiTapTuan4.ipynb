```python
#W4A10
n=int(input())
def collatz_length(k):
    length = 1
    while k != 1:
        if k % 2 == 0:
            k = k // 2
        else:
            k = 3 * k + 1
        length += 1
    return length
max_length = 0
min_x = 0
for i in range(1, n + 1):
    length = collatz_length(i)
    if length > max_length:
        max_length = length
        min_x = i
    elif length == max_length and i < min_x:
        min_x = i
print(min_x, max_length)
```

```python
#W4A13
a=int(input())
b=int(input())
tonga=0
tongb=0
for i in range(1,a):
    if a%i==0:
        tonga+=i
for j in range(1,b):
    if b%j==0:
        tongb+=j
if tonga==b and tongb==a:
    print("True")
else:
    print("False")
```
