```python
#W4A10
n=int(input())
def collatz_length(k):
    count=0
    while k!=1:
        if k%2!=0:
            k=3*k+1
            count+=1
        if k%2==0:
            k=k//2
            count+=1
    return count
min_length=0
collatz=0
for i in range(1,n+1):
    if collatz_length(i)>min_length:
        min_length=collatz_length(i)
        collatz=i
print(collatz)
print(min_length)
```
